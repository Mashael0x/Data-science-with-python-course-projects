### 5. Perform a statistical test for the following:



#### 1.Whether the average response time across complaint types is similar or not (overall)

```python
#NUll hypothesis: Average response time across complaint types is similar
 
#Alternate hypothesis: Average response time across complaint types is not similar
```

```python
top4_Complaints = data['Complaint Type'].value_counts()[:4]
top4_Complaints
```

```python
sample_data = data.loc[data['Complaint Type'].isin(top4_Complaints), ['Complaint Type', 'Request_Closing_Time']]
sample_data.head()
sample_data.dropna(how='any', inplace=True)
sample_data.isnull().sum()
```

```python
s1 = sample_data[sample_data['Complaint Type'] == top4_Complaints[0]].Request_Closing_Time
print(s1.head())
```

```python
s2 = sample_data[sample_data['Complaint Type'] == top4_Complaints[1]].Request_Closing_Time
print(s2.head())
```

```python
s3 = sample_data[sample_data['Complaint Type'] == top4_Complaints[2]].Request_Closing_Time
print(s3.head())
```

```python
s4 = sample_data[sample_data['Complaint Type'] == top4_Complaints[3]].Request_Closing_Time
print(s4.head())
```

```python
print(s1.isnull().sum())
print(s2.isnull().sum())
print(s3.isnull().sum())
print(s4.isnull().sum())
```

```python
print(stats.f_oneway(s1, s2, s3, s4))
```

### p-value is less than 0.05 so we reject null hypothesis, average response time is not similar.

#### 2.Are the type of complaint or service requested and location related?

```python
#NULL hypothesis: types of complaint or service requested and location are related.

#Alternate hypothesis: not related.
```

```python
chi_sq=pd.DataFrame()
chi_sq["Location Type"]=data["Location Type"]
chi_sq["Complaint Type"]=data["Complaint Type"]
```

```python
chi_sq.dropna(inplace=True)
```

```python
data_crosstab = pd.crosstab( chi_sq["Location Type"],chi_sq["Complaint Type"])
```

```python
from scipy.stats import chi2_contingency

stat, p, dof, expected = chi2_contingency(data_crosstab) 

alpha = 0.05
if p >= alpha: 
    print('unrelated (reject H0)') 
else: 
    print('related (H0 holds true)')
```
